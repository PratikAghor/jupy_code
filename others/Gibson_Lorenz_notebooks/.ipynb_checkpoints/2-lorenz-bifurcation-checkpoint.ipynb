# Bifurcations of Lorenz equations

## Bifurcations of the equilibria

Let EQ0 be the EQB at the origin, and EQ1 and EQ2 be the secondary EQBs 
that bifurcate off EQ0 at $\rho = 1$. 

```julia
using Plots
using LinearAlgebra
```

```julia
# Define functions for EQ1, the flow f(x), and the Jacobian Df(x)
# all parameterized by σ, β, ρ

EQ0 = [0.0; 0.0; 0.0]

function EQ1(σ, β, ρ) 
    if ρ <= 1
        return [0.0; 0.0; 0.0]
    else
        return [sqrt(β*(ρ-1)); sqrt(β*(ρ-1)); ρ-1]
    end
end

function lorenz(x, σ, β, ρ)
    dxdt = σ*(x[2] - x[1])
    dydt =  ρ*x[1] - x[2] - x[1]*x[3]
    dzdt = -β*x[3] + x[1]*x[2]
    
    [dxdt; dydt; dzdt]
end

function lorenz_Jacobian(x, σ, β, ρ)
    [ -σ      σ     0   ; 
     ρ-x[3]  -1   -x[1] ;
     x[2]    x[1]  -β   ]
end
```

```julia
# From the above, build functions that compute eigvals at origin and EQ1
σ = 10; β = 8/3

eq0eigs(ρ) = eigvals(lorenz_Jacobian(EQ0, σ, β, ρ))
eq1eigs(ρ) = eigvals(lorenz_Jacobian(EQ1(σ, β, ρ), σ, β, ρ))
```

```julia
ρ = 1.3455
#@show eq0eigs(ρ)
#@show eq1eigs(ρ)
eq1eigs(ρ)
```

```julia
# Make a plot of eigenvalues of the EQB at the origin
# using explicit formula for eigvals as function of σ, β, ρ

# set σ and β as constants
σ = 10; β = 8/3

# defined eigenvalues as functions of ρ
λ₁(ρ) = -β
λ₂(ρ) = (-(σ+1) - sqrt((σ+1)^2 + 4*σ*(ρ-1)))/2
λ₃(ρ) = (-(σ+1) + sqrt((σ+1)^2 + 4*σ*(ρ-1)))/2

# plot λ versus ρ
ρ = 0:0.1:30
plot(ρ, λ₁.(ρ), color=:red)
plot!(ρ, λ₂.(ρ), color=:blue)
plot!(ρ, λ₃.(ρ), color=:green)
plot!(ρ, 0*ρ, color=:black, linestyle=:dash)

plot!(xlabel="rho", ylabel="lambda", legend=false)
plot!(title="eigenvalues of EQ0 = origin")
```

```julia
# Plot the eigenvalues of EQ1 in the complex plane

# Choose unusual sequence for ρ to resolve λs near bifurcation point
ρ = vcat(1.02:0.01:1.36, 2:2:30)
N = length(ρ)

# Λ = real and imag parts of three eigvals as func of ρ
Λ = zeros(N,6)

for n in 1:N
    λ = eq1eigs(ρ[n])
    for j = 0:2
        Λ[n,2j+1] = real(λ[j+1])
        Λ[n,2j+2] = imag(λ[j+1])
    end
end

mrk = :circ
plot(Λ[:,1], Λ[:,2], marker=mrk, markersize=2)
plot!(Λ[:,3], Λ[:,4], marker=mrk, markersize=2)
plot!(Λ[:,5], Λ[:,6], marker=mrk, markersize=2)
plot!(xlabel="Re lambda", ylabel="Im lambda", legend=false)
plot!(title="eigvals of EQ1")
plot!(xlim=(-15,1))
```

## The effect of bifurcations on trajectories

Here we'll plot trajectories of the Lorenz system for $\sigma = 10, \beta =8/3$, and a variety of values of $\rho$, 
specifically the unstable manifold of the origin EQ0 (obtained fin order to observe the change of system behavior through the equilibrium bifurcations, and also the global bifurcation that occurs as we pass through a homoclinic orbit. The trajectories 


Key points:

$\rho = 1$ bifurcation of EQ1,2 off EQ0 (origin)

$\rho \approx 1.3456$ a stable real eigenvalue of EQ1, EQ2 goes complex

$\rho \approx 13.926$ homoclinic orbit -global bifurcation

$\rho \approx 24.74$ complex eigenvalue of EQ1, EQ2 goes unstable

```julia
function rungekutta4(f, x₀, Δt, T) 
   
    # fix up Δt so it evenly divides time interval
    N = convert(Int64, round(T/Δt))
    #Δt = (t₁ - t₀)/N
    Δt2 = Δt/2
    Δt6 = Δt/6
    t = (0:N)*Δt

    x = zeros(N+1, length(x₀))
    x[1,:] = x₀

    for n = 1:N
        xn = x[n,:]
        tn = t[n]
        s1 = f(xn)
        s2 = f(xn + Δt2*s1)
        s3 = f(xn + Δt2*s2)
        s4 = f(xn + Δt *s3)

        x[n+1, :] =  xn + Δt6*(s1 + 2s2 + 2s3 + s4) # RK4 update formula
    end
    t,x
end
```

```julia
function lorenzplot(σ, β, ρ, T)
    
    # equilibrium points
    eq0 = [0.0; 0.0; 0.0]
    eq1 = copy(eq0)
    eq2 = copy(eq0)
    
    if ρ > 1
        c = sqrt(β*(ρ-1))
        eq1 = [-c; -c; ρ-1]
        eq2 = [ c;  c; ρ-1]
    end

    ϵ = 1e-02
    Δt = 0.01

    # construct f(x) and Df(x) functions for given vals of σ, β, ρ
    f(x) = lorenz(x, σ, β, ρ)
    Df(x) = lorenz_Jacobian(x, σ, β, ρ)

    # compute leading eigenvector of Df at origin
    Λ,V = eigen(Df(eq0))
    p = sortperm(real(Λ), rev=true) # find largest eigval
    v = V[:,p[1]]
    
    # plot the eqbs
    plot([eq0[1]], [eq0[2]], [eq0[3]], marker=:circ, color=:black)
    plot!([eq1[1]], [eq1[2]], [eq1[3]], marker=:circ, color=:black)
    plot!([eq2[1]], [eq2[2]], [eq2[3]], marker=:circ, color=:black)
    
    # plot trajectories in unstable manifold of origin
    x₀ = eq0 - ϵ*v
    t,x = rungekutta4(f, x₀, Δt, T)
    plot!(x[:,1], x[:,2], x[:,3], color=:blue)
    
    x₀ = eq0 + ϵ*v
    t,x = rungekutta4(f, x₀, Δt, T)
    plot!(x[:,1], x[:,2], x[:,3], color=:red)
    plot!(legend=false, xlabel="x, y", ylabel="z")
    #plot!(xlim=(-30,30), ylim=(-30,30), zlim=(0,60))
end
```

```julia
lorenzplot(10, 8/3, 24.74, 30)
plot!(xlim=(-30,30), ylim=(-30,30), zlim=(0,60))
```
